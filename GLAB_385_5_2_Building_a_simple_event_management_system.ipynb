# **Lab Overview:**
Here's a real-world example of using Python's date and time functionality to build a simple event management system. In this example, we'll create a program that allows users to schedule and list events, and It displays upcoming events and shows how many days remain until each one..

- **Build a simple event management system using Python.**
- **Use Python’s `datetime` module** to work with dates and perform date calculations.
- **Store events in a dictionary named `events`:**
  - **Key:** Event name  
  - **Value:** Event date
- **Allow users to add events** by entering:
  - An **event name**
  - An **event date in the format `YYYY-MM-DD`**
- **Validate the date format** to ensure correct user input.
- **Display a list of upcoming events** when the user selects the list option.
- **Sort events by date** so the earliest events appear first.
- **Calculate and display the number of days remaining** until each event.
- **Run the program continuously in a menu-driven loop** until the user chooses to quit.

```python
import datetime  # Allows us to work with dates

# Dictionary to store events
# Key = event name
# Value = event date
events = {}


# Function to add a new event
def add_event():

    # Ask the user to enter the event name
    event_name = input("Enter the event name:  ")

    # Ask the user to enter the event date
    date_input = input("Enter the event date (YYYY-MM-DD):  ")

    # Try converting the input string into a date
    try:
        event_date = datetime.datetime.strptime(date_input, "%Y-%m-%d").date()
    except ValueError:
        # If the date format is incorrect
        print("Invalid date format. Please use YYYY-MM-DD.")
        return

    # Save the event in the dictionary
    events[event_name] = event_date

    # Confirmation message
    print("Event successfully added!")


# Function used to help sort events by date
def get_event_date(item):
    # item looks like: ("Event Name", event_date)
    return item[1]  # Return only the date for sorting


# Function to display events
def list_events():

    # Check if the dictionary is empty
    if len(events) == 0:
        print("No upcoming events.")
        return

    print("\nUpcoming Events:")

    # Sort events by date
    sorted_events = sorted(events.items(), key=get_event_date)

    # Go through each event in the sorted list
    for event_name, event_date in sorted_events:

        # Calculate the number of days until the event
        today = datetime.date.today()
        days_remaining = (event_date - today).days

        # Display the event information
        print(event_name, "-", event_date, "-", days_remaining, "days remaining")


# Main menu function
def main():

    while True:

        # Display the menu
        print("\nEvent Management System")
        print("1. Add Event")
        print("2. List Events")
        print("3. Quit")

        # Get the user's choice
        choice = input("Enter your choice: ")

        # Call the appropriate function
        if choice == "1":
            add_event()

        elif choice == "2":
            list_events()

        elif choice == "3":
            print("Program ended.")
            break

        else:
            print("Invalid option. Please try again.")


# Start the program
if __name__ == "__main__":
    main()
```

# **Explaination:**

Let's go through the program step by step.

### Creating an Empty Dictionary for Events
- The program creates an empty dictionary named **`events`** to store event names and their corresponding dates.
- This dictionary acts as the program’s **data store**.
- **Key:** Event name  
- **Value:** Event date

### `add_event()` Function
- The **`add_event()`** function allows users to add events to the system.
- It prompts the user to enter:
  - An **event name**
  - An **event date in the format `YYYY-MM-DD`**
- The program attempts to convert the date string into a **`datetime.date` object** using `strptime(...).date()`.
- If the user enters an invalid date format, the program displays an **error message** and does not add the event.

### `list_events()` Function
- The **`list_events()`** function displays a list of upcoming events.
- Events are **sorted by date** so the earliest event appears first.
- The program calculates the **number of days remaining** until each event.
- If there are no events stored in the dictionary, the program informs the user that **no upcoming events exist**.

### `main()` Function
- The **`main()`** function acts as the **control center** of the program.
- It displays a menu with three options:
  1. Add Event  
  2. List Events  
  3. Quit
- A **`while` loop** keeps the program running so the user can continue interacting with it until they choose to quit.

### Running the Program
- The condition **`if __name__ == "__main__":`** ensures that the **`main()`** function runs only when the script is executed directly.
- The user can repeatedly add events, list events, or exit the program.
- The program **continues running until the user selects the quit option**.
**Running the Program:**

The if __name__ == "__main__": condition ensures that the main() function is executed when the script is run.
Users can add events, list events, or quit the program. The program will keep running until the user chooses to quit.
